### 4. Створити декілька власних об’єктів DataFrame за такою ж тематикою, що й файл. Наприклад, якщо тема файлу – жаби, можна створити об’єкти, що містять розміри жаб, вагу, стать, кількість особин в популяції і т.д. Використати описані в теоретичних відомостях параметри методів merge та concat для різних видів злиття та об’єднання даних цих об’єктів.

```python
# Один-до-одного
a = pd.DataFrame({'mage': [20, 19, 32, 25], 'Birthweight': [2.4, 4.2, 3.3, 2.7]})
b = pd.DataFrame({'mage': [20, 19, 32, 25], 'smoker': [1, 0, 1, 1]}) 
c = pd.merge(a, b)
c
```

```python
# Один-до-одного, але назва стовпчика інша
a = pd.DataFrame({'mage': [20, 19, 32, 25], 'Birthweight': [2.4, 4.2, 3.3, 2.7]})
b = pd.DataFrame({'mother_age': [20, 19, 32, 25], 'smoker': [1, 0, 1, 1]}) 
c = pd.merge(a, b, left_on="mage", right_on="mother_age").drop('mother_age', axis = 1)
c
```

```python
# Багато-до-одного
a = pd.DataFrame({'mage': [20, 19, 32, 25], 'smoker': [1, 0, 1, 1], 'Birthweight': [2.4, 4.2, 3.3, 2.7]})
b = pd.DataFrame({'mage': [20, 19, 32, 25], 'fage': [24, 18, 36, 30]}) 
c = pd.merge(a, b)
c
```

```python
# Багато-до-багатьох
a = pd.DataFrame({'mage': [20, 19, 32, 25], 'smoker': [1, 0, 1, 1], 'Birthweight': [2.4, 4.2, 3.3, 2.7]})
b = pd.DataFrame({'mage': [20, 19, 32, 25], 'fage': [24, 18, 36, 30], 'Length': [48, 50, 53, 49]}) 
c = pd.merge(a, b)
c
```

```python
a = pd.DataFrame({'mage': [20, 19, 32, 25], 'smoker': [1, 0, 1, 1]})
b = pd.DataFrame({'mage': [21, 19, 33, 26], 'fage': [24, 18, 36, 30]}) 
c = pd.merge(a, b)
c
```

```python
a = pd.DataFrame({'mage': [20, 19, 32, 25], 'smoker': [1, 0, 1, 1]})
b = pd.DataFrame({'mage': [21, 19, 33, 26], 'fage': [24, 18, 36, 30]}) 
c = pd.merge(a, b, how = 'outer')
c
```

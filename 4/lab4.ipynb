```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_csv('diamonds.csv')
data
```

### 1. Побудувати стовпчикові діаграми, на яких відобразити:
#### а) кількість діамантів кожного з класів якості; 

```python
sns.countplot(data = data, x = 'cut', palette = 'RdPu')
#data['cut'].value_counts().plot(kind = 'bar'); #або barh
```

#### б) максимальну ціну діамантів кожного класу якості; 

```python
max = pd.pivot_table(data, values = 'price', index = 'cut', aggfunc = np.max)
max.plot(kind = 'bar')
```

#### в) середню глибину діамантів різного класу якості з різною якістю кольору.

```python
sns.barplot(data = data, x = 'cut', y = 'depth', hue = 'color', palette = 'RdPu')
```

### 2. Побудувати гістограму глибини діамантів у відсотках (depth), загальну і для кожного класу якості.

```python
sns.histplot(data = data, x = 'depth', palette = 'RdPu') # bins задає кількість стовпців
#plt.hist(data = data, x = 'depth')
```

```python
sns.histplot(data = data, x = 'depth', hue = 'cut', multiple = 'stack', palette = 'RdPu')
```

### 3. Побудувати діаграму розмаху параметру table (загальну і в залежності від якості кольору), визначити чи присутні викиди.


```python
sns.boxplot(data = data, y = 'table', palette = 'RdPu')
```

```python
sns.boxplot(data = data, y = 'table', x = 'color', palette = 'RdPu')
```
